````python
import sys

import requests
from bs4 import BeautifulSoup
import re
import unicodedata
import pandas as pd
````

and we will provide some helper functions for you to process web scraped HTML table


````python
def date_time(table_cells):
    """
    This function returns the data and time from the HTML  table cell
    Input: the  element of a table data cell extracts extra row
    """
    return [data_time.strip() for data_time in list(table_cells.strings)][0:2]

def booster_version(table_cells):
    """
    This function returns the booster version from the HTML  table cell 
    Input: the  element of a table data cell extracts extra row
    """
    out=''.join([booster_version for i,booster_version in enumerate( table_cells.strings) if i%2==0][0:-1])
    return out

def landing_status(table_cells):
    """
    This function returns the landing status from the HTML table cell 
    Input: the  element of a table data cell extracts extra row
    """
    out=[i for i in table_cells.strings][0]
    return out


def get_mass(table_cells):
    mass=unicodedata.normalize("NFKD", table_cells.text).strip()
    if mass:
        mass.find("kg")
        new_mass=mass[0:mass.find("kg")+2]
    else:
        new_mass=0
    return new_mass


def extract_column_from_header(row):
    """
    This function returns the landing status from the HTML table cell 
    Input: the  element of a table data cell extracts extra row
    """
    if (row.br):
        row.br.extract()
    if row.a:
        row.a.extract()
    if row.sup:
        row.sup.extract()
        
    colunm_name = ' '.join(row.contents)
    
    # Filter the digit and empty names
    if not(colunm_name.strip().isdigit()):
        colunm_name = colunm_name.strip()
        return colunm_name    
````

### TASK 2: Extract all column/variable names from the HTML table header


Next, we want to collect all relevant column names from the HTML table header


Let's try to find all tables on the wiki page first. If you need to refresh your memory about `BeautifulSoup`, please check the external reference link towards the end of this lab


````python
# Use the find_all function in the BeautifulSoup object, with element type `table`
# Assign the result to a list called `html_tables`
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Falcon_9"
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")
    # Find the table element containing the column names
    table = soup.find("table", class_="wikitable")
    # Extracting header row
    header_row = table.find("tr")
    # Extracting column names from header row
    column_names = [header.text.strip() for header in header_row.find_all("th")]
    
    print("Column Names:")
    for column_name in column_names:
        print(column_name)
else:
    print("Failed to retrieve the page. Status code:", response.status_code)

````

Next, we just need to iterate through the `<th>` elements and apply the provided `extract_column_from_header()` to extract column name one by one


````python
column_names = []
import requests
from bs4 import BeautifulSoup

# Function to extract column name from header
def extract_column_from_header(header):
    return header.text.strip()

url = "https://en.wikipedia.org/wiki/Falcon_9"
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")
    # Find all tables on the page
    html_tables = soup.find_all("table", class_="wikitable")
    # Check if at least three tables are found
    if len(html_tables) >= 3:
        # Select the third table (index 2 since Python is zero-indexed)
        target_table = html_tables[2]
        # Extracting column names from the header row of the third table
        header_row = target_table.find("tr")
        # Iterating through <th> elements and applying extract_column_from_header function
        for header in header_row.find_all("th"):
            column_names.append(extract_column_from_header(header))
        
        print("Column Names:")
        for column_name in column_names:
            print(column_name)
    else:
        print("There are less than three tables on the page.")
else:
    print("Failed to retrieve the page. Status code:", response.status_code)


# Apply find_all() function with `th` element on first_launch_table
# Iterate each th element and apply the provided extract_column_from_header() to get a column name
# Append the Non-empty column name (`if name is not None and len(name) > 0`) into a list called column_names
````
